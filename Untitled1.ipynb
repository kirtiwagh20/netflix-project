```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

# **DATA LOADING**

```python
data_df = pd.read_csv('netflix1 (1).csv')
display(data_df.head())
print(data_df.shape)
```

# **DATA CLEANING**

```python
print(data_df.isnull().sum())
data_df.drop_duplicates(inplace=True)
data_df.dropna(subset = ['director', 'country'], inplace=True)
```

```python
data_df['date_added'] = pd.to_datetime(data_df['date_added'])
data_df
```

# **DATA ANALYSIS**

```python
type_count = data_df['type'].value_counts()
print(type_count)
```

```python
plt.figure(figsize=(8, 6))
plt.pie(type_count.values, labels=type_count.index, autopct='%1.1f%%')
plt.title("Distribution of Content Types on Netflix")
plt.show()
```

```python
data_df['genres'] = data_df['listed_in'].str.split(', ')
data_df = data_df.explode('genres')
genre_count = data_df['genres'].value_counts()
print(genre_count)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(genre_count.index, genre_count.values)
plt.xlabel('Genres')
plt.ylabel('Count')
plt.title('Distribution of Genres on Netflix')
plt.xticks(rotation=45)
plt.show()
```

```python
data_df['release_year'] = data_df['date_added'].dt.year
year_count = data_df['release_year'].value_counts()
print(year_count)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(year_count.index, year_count.values, color='lightgreen')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.title('Distribution of Release Years on Netflix')
```

```python
filtered_data = data_df[data_df['director'] != 'Not Given']
top_directors = filtered_data['director'].value_counts().head(10)
print(top_directors)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top_directors.index, top_directors.values, color='skyblue')
plt.xlabel('Director')
plt.ylabel('Count')
plt.title('Top 10 Directors with the Most Content on Netflix')
```

```python
yearly_movie_release = data_df[data_df['type'] == 'Movie']['release_year'].value_counts().sort_index()
yearly_tv_show_release = data_df[data_df['type'] == 'TV Show']['release_year'].value_counts().sort_index()
print(yearly_movie_release)
print(yearly_tv_show_release)
```

```python
plt.plot(yearly_movie_release.index, yearly_movie_release.values, label='Movies')
plt.plot(yearly_tv_show_release.index, yearly_tv_show_release.values, label='TV Shows')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Yearly Release of Movies and TV Shows on Netflix')
```

```python

text = " ".join(genre for genre in data_df.genres)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Genres on Netflix')
plt.show()
```

```python

import ipywidgets as widgets
from IPython.display import display, clear_output

genre_dropdown = widgets.Dropdown(
    options=data_df['listed_in'].unique(),
    value=data_df['listed_in'].unique()[0],
    description='Select Genre:'
)

output = widgets.Output()

def update_plot(change):
    selected_genre = change.new
    with output:
        clear_output(wait=True)
        genre_data = data_df[data_df['listed_in'] == selected_genre]
        if not genre_data.empty:
            plt.figure(figsize=(8, 6))
            sns.countplot(x='release_year', data=genre_data)
            plt.xlabel('Release Year')
            plt.ylabel('Count')
            plt.title(f'Release Year Distribution for {selected_genre}')
            plt.xticks(rotation=45)
            plt.show()
        else:
            print(f"No data found for genre '{selected_genre}'")


genre_dropdown.observe(update_plot, names='value')

# Display the interactive dashboard
display(genre_dropdown)
display(output)
```
